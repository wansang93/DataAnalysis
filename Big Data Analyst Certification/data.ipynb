```python
import pandas as pd
```

```python
LINK = './mtcars.csv'
df = pd.read_csv(LINK)
df.head()
```

```python
# 1사분위, 2자리 반올림
q1 = df['mpg'].quantile(0.25)
print(round(q1, 2))
```

```python
# 조건문 19 ~ 21
cond1 = df['mpg'] >= 19
cond2 = df['mpg'] <= 21
df[~ cond1 & cond2][['model', 'mpg']]
```

```python
# IQR 구하기
Q3 = df['hp'].quantile(0.75)
Q1 = df['hp'].quantile(0.25)
IQR = Q3 - Q1
print(IQR)
```

```python
# wt 변수의 상위 10개 값의 총 합
top10 = df.sort_values('wt', ascending=False)
sum_top10 = sum(top10['wt'].head(10))
print(int(sum_top10))
```

```python
# 전체 차 중 cyl의 6의 비율, 첫째자리
cond = df['cyl'] == 6
val = len(df[cond]['cyl'])
total = len(df)

print(round( val / total, 1))
```

```python
# 10개 mpg 열 평균값 반올림 정수
mean10 = df['mpg'].loc[:9].mean()
round(mean10)
```

```python
LINK = "https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv"
df = pd.read_csv(LINK, index_col=0)
df.head()
```

```python
# channelTitle(변경가능), channelId(변경불가) 인데 title 바꾼 채널 갯수
change_count = df.groupby('channelId')['channelTitle']
```
