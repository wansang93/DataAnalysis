```python
import pandas as pd
import os
encoding = 'latin1'
```

```python
upath = os.path.expanduser('./data/2.2 movielens/ml-1m/users.dat')
upath
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
unames
```

```python
users = pd.read_csv(upath, sep='::', header=None, names=unames, encoding=encoding, engine='python')
users
```

```python
rpath = os.path.expanduser('./data/2.2 movielens/ml-1m/ratings.dat')
mpath = os.path.expanduser('./data/2.2 movielens/ml-1m/movies.dat')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']

ratings = pd.read_csv(rpath, sep='::', header=None, names=rnames, encoding=encoding, engine='python')
movies = pd.read_csv(mpath, sep='::', header=None, names=mnames, encoding=encoding, engine='python')
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
ratings
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
top_female_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff
```

```python
sorted_by_diff[::-1][:15]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title
```

```python
rating_std_by_title = rating_std_by_title.ix[active_titles]
rating_std_by_title
```

```python
rating_std_by_title.sort_values(ascending=False)[:10]
```
